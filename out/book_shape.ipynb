```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../final_shape.csv")
```

```python
df
```

```python
plt.bar(df["delta_ticks"], df["bid_qty"], width=10, color="green", alpha=0.6, label="Bids")
plt.bar(df["delta_ticks"], -df["ask_qty"], width=10, color="red", alpha=0.6, label="Asks")
plt.xlabel("Distance from mid (ticks)")
plt.ylabel("Quantity")
plt.title("Orderbook Shape")
plt.legend()
plt.show()
```

```python
df = pd.read_csv("../final_shape.csv").sort_values("delta_ticks")
plt.figure(figsize=(8,5))
plt.bar(df["delta_ticks"], df["bid_qty"], width=1, color="green", alpha=0.6, label="Bids")
plt.bar(df["delta_ticks"], -df["ask_qty"], width=1, color="red", alpha=0.6, label="Asks")
plt.xlim(-1000, 1000)  # zoom near mid
plt.xlabel("Distance from mid (ticks)")
plt.ylabel("Quantity")
plt.title("Orderbook Shape")
plt.legend()
plt.show()
```
